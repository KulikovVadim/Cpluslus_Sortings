```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [10, 100, 1000, 10000, 100000]
y = [0.00096, 0.00525, 5.86742, 561.985, 46045.43]
fig, ax = plt.subplots ()
ax.plot (x,y, marker='o')
ax.set_title ('duration of bubbl–µ sort')
ax.set_xlabel ("elements")
ax.set_ylabel ("time, ms")

plt.show
```
